```julia
# Metodo 1: devolver como tupla
function multCmplj(operando1,operando2)
    prodReal = operando1[1]*operando2[1]-operando1[2]*operando2[2]
    prodImaginario = operando1[1]*operando2[2]+operando1[2]*operando2[1]
    return prodReal, prodImaginario
end
```

```julia
function modulo(complejo)
    r_2 = complejo[1]*complejo[1]
    i_2 = complejo[2]*complejo[2]
    resultado = sqrt(r_2 + i_2)
    return resultado
end
```

```julia
# Metodo 2 : devolver como vector
function multCmplj(operando1,operando2)
    prodReal = operando1[1]*operando2[1]-operando1[2]*operando2[2]
    prodImaginario = operando1[1]*operando2[2]+operando1[2]*operando2[1]
    return [prodReal, prodImaginario]
end
```

```julia
x = range(-2,2,length=10)
y = range(-2,2,length=10)
for i in 1:10
    for j in 1:10
        a = x[i]
        b = y[j]
        dato = [a,b]
        A[i,j] = modulo(dato)
    end
    println("")
end

println(A)
```

```julia
function divergeQ(z)
    if modulo(z)>2
        return true
    else
        return false
    end
end
```

```julia
n = 1000
fractal = 127*ones(n,n)

x = range(-2,1.5,length=n)
y = range(-1.5,1.5,length=n)
for i in 1:n
    for j in 1:n
        a = x[i]
        b = y[j]
        dato = [a,b]
        z = [0,0]
        for k in 1:256
            z = multCmplj(z,z) + dato
            if divergeQ(z)
                fractal[i,j]=modulo(z)
                break
            end
        end
    end
end
```

```julia
mandel = heatmap(fractal, color=:blues)
#savefig(mandel, "mandelbrot_mayora2.png")
```
