```julia
function productoPunto(U1,U2)
    n1 = length(U1)
    n2 = length(U2)
    resultado = 0.0
    if n1 == n2
        for i=1:n1
            resultado = resultado + U1[i]*U2[i]
        end
    else
        println("Tus vectores deben tener el mismo tamao, intenta de nuevo")
        return NaN
    end
    return resultado
end

v1 = [1,-1,3]
v2 = [2,-3,1,0]
println(productoPunto(v1,v2))
```

```julia
#=
Av
= A(,i) * vi ) (a_i*b_i)
=#

function matrizPorVector(M,V)
    renglones, columnas = size(M)
    n = length(V)
    resultado = zeros(renglones)
    for j = 1:renglones, i=1:columnas
        resultado[j] = resultado[j] + M[j,i]*V[i]
    end
    return [resultado]
end

Matriz = ones(3,4)
Vector = ones(4)

matrizPorVector(Matriz,Vector)
```

```julia
function MatrizPorMatriz(A,B)
    renglones_a, columnas_a = size(A)
    renglones_b, columnas_b = size(B)
    C = zeros(renglones_a, columnas_b)
    for i=1:renglones_a, j=1:columnas_b, k=1:columnas_a
        C[i,j] = C[i,j] + A[i,k]*B[k,j]
    end
    return C
end
```

```julia
Mat1=ones(2,3)
Mat2=ones(3,2)
println(MatrizPorMatriz(Mat1,Mat2))
```
